# Raw Property Data (Table View)
Source: `data/raw_data/property.json`


```python
import os
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# Resolve repo root whether notebook lives at repo root or notebooks/
NOTEBOOK_DIR = os.path.abspath('')
if os.path.exists(os.path.join(NOTEBOOK_DIR, 'config.py')):
    PROJECT_DIR = NOTEBOOK_DIR
else:
    PROJECT_DIR = os.path.dirname(NOTEBOOK_DIR)

APP_NAME = 'RawPropertyData'
DATA_PATH = os.path.join(PROJECT_DIR, 'data/raw_data/property.json')

spark = (
    SparkSession.builder
    .appName(APP_NAME)
    .master('local[*]')
    .getOrCreate()
)
spark.sparkContext.setLogLevel('WARN')

df_raw = spark.read.option('multiLine', 'true').json(DATA_PATH)

df_view = df_raw.select(
    F.col('id').alias('id'),
    F.coalesce(F.col('name.`en-us`'), F.col('name.de'), F.col('name.fr'), F.col('name.es')).alias('name'),
    F.col('location.country').alias('country'),
    F.col('location.city').alias('city_id'),
    F.col('currency').alias('currency'),
    F.col('rating.stars').alias('stars'),
    F.col('rating.review_score').alias('review_score'),
    F.col('accommodation_type').alias('accommodation_type'),
    F.col('number_of_rooms').alias('number_of_rooms'),
    F.col('is_genius').alias('is_genius'),
    F.col('is_work_friendly').alias('is_work_friendly'),
    F.col('url').alias('url')
)

df_view.show(20, truncate=False)

spark.createDataFrame(df_view.dtypes, ['column', 'dtype']).show(truncate=False)

df_view.count()
```
