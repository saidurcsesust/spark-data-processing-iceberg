# Raw Search Data (Table View)
Source: `data/raw_data/search.json`


```python
import os
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# Resolve repo root whether notebook lives at repo root or notebooks/
NOTEBOOK_DIR = os.path.abspath('')
if os.path.exists(os.path.join(NOTEBOOK_DIR, 'config.py')):
    PROJECT_DIR = NOTEBOOK_DIR
else:
    PROJECT_DIR = os.path.dirname(NOTEBOOK_DIR)

APP_NAME = 'RawSearchData'
DATA_PATH = os.path.join(PROJECT_DIR, 'data/raw_data/search.json')

spark = (
    SparkSession.builder
    .appName(APP_NAME)
    .master('local[*]')
    .getOrCreate()
)
spark.sparkContext.setLogLevel('WARN')

df_raw = spark.read.option('multiLine', 'true').json(DATA_PATH)

df_view = df_raw.select(
    F.col('id').alias('id'),
    F.col('currency').alias('currency'),
    F.col('price.book').alias('price_book'),
    F.col('commission.percentage').alias('commission_pct'),
    F.col('deep_link_url').alias('deep_link_url'),
    F.col('url').alias('url'),
    F.size('products').alias('products_count')
)

df_view.show(20, truncate=False)

spark.createDataFrame(df_view.dtypes, ['column', 'dtype']).show(truncate=False)

df_view.count()
```
